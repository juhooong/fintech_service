```python
def ages(x):
    if 10 <= x < 20:
        return "10s"
    elif 20 <= x < 30:
        return "20s"
    elif 30 <= x < 40:
        return "30s"
    elif 40 <= x < 50:
        return "40s"
    elif 50 <= x < 60:
        return "50s"
    elif 60 <= x < 70:
        return "60s"
    elif 70 <= x < 80:
        return "70s"
    elif 80 <= x < 90:
        return "80s"
    elif 90 <= x < 100:
        return "90s"
```

```python
data['age'] = data['age'].apply(ages)
```

```python
def workclasses(x):
    if 'gov' in x:
        return "G"
    elif 'Self' in x:
        return "S"
    elif x == 'Private':
        return "P"
    elif x == 'Without-pay':
        return "W"
```

```python
data['workclass'] = data['workclass'].apply(workclasses)
```

```python
data['education'] = data['education'].str.strip()
```

```python
def educations(x):
    if x == 'Preschool':
        return 'P'
    elif (x == '1st-4th') | (x == '5th-6th'):
        return "E"
    elif (x == '7th-8th') | (x == '9th'):
        return "M"
    elif (x == '10th') | (x == '11th') | (x == '12th') | (x == 'HS-grad'):
        return "H"
    elif (x == 'Some-college') | (x == 'Assoc-acdm') | (x =='Assoc-voc') | (x == 'Bachelors'):
        return "U"
    elif (x == 'Masters') | (x == 'Doctorate') | (x ==  'Prof-school'):
        return "G"
```

```python
data['education'] = data['education'].apply(educations)
```

```python
def occupations(x):
    if x == 'Farming-fishing':
        return 'F'
    elif (x == 'Armed-Forces') | (x == 'Protective-serv'):
        return "P"
    elif (x == 'Prof-specialty') | (x == 'Exec-managerial') | (x == 'Adm-clerical') | (x == 'Tech-support') | (x == 'Sales'):
        return "W"
    else:
        return "B"
```

```python
data['occupation'] = data['occupation'].apply(occupations)
```
