# 배열 합치기
* np.concatenate()

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
print(a)
print(a.shape)
print(b)
print(b.shape)
```

* np.concatenate((a,b)) 의 기본 병합 방향은 행방향, axis = 0
* 행방향으로 합칠 때는 두 행렬의 열의 길이가 같아야 합쳐짐

```python
concat_row = np.concatenate((a,b))
print(concat_row)
print(concat_row.shape)
```

* axis=1은 열방향으로 붙음
* 두 행렬의 행 길이가 같아야 붙음

```python
concat_col = np.concatenate((a,b), axis=1)
print(concat_col)
print(concat_col.shape)
```

# 합치는 방향의 모양이 일치하지 않으면 합치기 X

```python
c = np.array([[1,2],[3,4]])
d = np.array([[5,6]])
print(c)
print(c.shape)
print(d)
print(d.shape)
```

```python
concat_row2 = np.concatenate((c,d), axis=0)
print(concat_row2)
print(concat_row2.shape)
```

```python
concat_col2 = np.concatenate((c,d.T), axis=1)
print(concat_col2)
print(concat_col2.shape)
```

## 벡터와 매트릭스는 합칠 수 없다.

```python
e = np.array([[1,2],[3,4]])
f = np.array([5,6])
print(e)
print(e.shape)
print(e.ndim)
print(f)
print(f.shape)
print(f.ndim)
```

```python
concat_row3 = np.concatenate((e,f), axis=0)
print(concat_row3)
print(concat_row3.shape)
```

# 전치 행렬 만들기
* .T
* .transpose()
* .reshape()

```python
h = np.array([[5,6,7]])
```

```python
print(h)
print(h.shape)
```

```python
h2 = h.T
print(h2)
print(h2.shape)
```

```python
h3 = h.transpose()
print(h3)
print(h3.shape)
```

```python
h4 = h.reshape(3,1)
print(h4)
print(h4.shape)
```

```python
i = np.arange(6).reshape(3,2)
print(i)
print(i.shape)
print(i.ndim)
```

```python
i2 = i.T
print(i2)
print(i2.shape)
```

```python
i3 = i.transpose()
print(i3)
print(i3.shape)
```

```python
i4 = i.reshape(2,3)
print(i4)
print(i4.shape)
```
