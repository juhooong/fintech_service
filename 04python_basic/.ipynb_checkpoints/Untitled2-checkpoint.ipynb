```python
class FourCal():
    def setdata(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
        print("self.num1 :", self.num1)
        print("self.num2 :", self.num2)
        
    def add(self):
        result = self.num1 + self.num2
        return result
    
    def sub(self):
        result = self.num1 - self.num2
        return result
    
    def mul(self):
        result = self.num1 * self.num2
        return result
    
    def div(self):
        result = self.num1 / self.num2
        return result
```

```python
sam.mul()
```

```python
sam.div()
```

```python
class FourCal2():
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
        
    def add(self):
        result = self.num1 + self.num2
        return result
        
    def sub(self):
        result = self.num1 - self.num2
        return result
        
    def mul(self):
        result = self.num1 * self.num2
        return result
    
    def div(self):
        result = self.num1 / self.num2
        return result
```

```python
james = FourCal2(100, 27)
```

```python
james.add()
```

```python
james.sub()
```

```python
james.mul()
```

```python
james.sub()
```

```python
class MoreFourCal(FourCal2):
    def pow(self):
        result = self.num1 ** self.num2
        return result
```

```python
sally = MoreFourCal(100, 4)
```

```python
sally.pow()
```

```python
sally.add()
```

```python
sally.sub()
```

```python
class SafeFourCal(FourCal2):
    def div(self):
        if self.num2 == 0:
            return 0
        else:
            return self.num1 / self.num2
```

```python
john = SafeFourCal(4, 0)
```

```python
john.div()
```
