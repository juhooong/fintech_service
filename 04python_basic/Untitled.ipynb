```python
def even(age):
    for i in age:
        if i % 2 == 0:
            print(i, end = " ")
        elif i >= 7:
            break
```

```python
even(age)
```

```python
age = range(1, 101)
```

```python
def odd(age):
    for i in age:
        if i % 2 != 0:
            print(i, end = " ")
        elif i >= 51:
            return
```

```python
odd(age)
```

```python
def muls(a, *args):
    result = []
    for num in args:
        print(a * num, end=' ')
        result.append(a * num)
    return result
```
