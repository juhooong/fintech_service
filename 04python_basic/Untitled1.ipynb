```python
kor = int(input("국어 점수: "))
eng = int(input("영어 점수: "))
math = int(input("수학 점수: "))
sci = int(input("과학 점수: "))
avg = (kor + eng + math + sci) / 4

if 0 <= kor <= 100 and 0 <= eng <= 100 and 0 <= math <= 100 and 0 <= sci <= 100:
    if avg >= 80:
        print('과락없이 합격입니다.', '평균:', avg)
    elif 60<= avg < 80 :
        print('과락은 없지만 평균 80점 미만으로 불합격입니다.', '평균:', avg)

    else:
        print('60점 미만인 과목이 있어 과락으로 불합격 입니다')
else:
    print('0-100 사이의 점수를 입력해 주세요.')
```
